```python
# importing libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
#import dataset
data=pd.read_csv(r"C:\Users\Gaonkar\Desktop\Ml+Dl+Rl_Projects\Ml\Sparks_Foundation\Task2 unsupervised\Iris.csv")
```

```python
data.head()
```

```python
data["Species"].unique()
```

```python
# replacing object type to int using replace function
data["Species"].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}, inplace=True)
```

```python
data=data.sample(frac=1,random_state=42)## shuffling the data seems prior data is unshuffles from first five rows
```

```python
features=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

```python
## Its true that in unsupervised learning we dont know the label but in this case we have the label so lets use it .
## we can use this label for checking our model form the ground truth.
## these are possible plot with the true label
for x1 in features:
    for x2 in features:
        if x1!=x2:
            fig, ax = plt.subplots()
            ax.scatter(data[data["Species"]==0][x1],data[data["Species"]==0][x2] , c='blue', label='Iris-setosa',alpha=0.3, edgecolors='none')
            ax.scatter(data[data["Species"]==1][x1],data[data["Species"]==1][x2] , c='orange', label='Iris-versicolor',alpha=0.3, edgecolors='none')
            ax.scatter(data[data["Species"]==2][x1],data[data["Species"]==1][x2] , c='green', label='Iris-virginica',alpha=0.3, edgecolors='none')
            ax.set_xlabel(x1)
            ax.set_ylabel(x2)
            ax.legend()
            ax.grid(True)

            plt.show()
```

```python
## we will use soft clustering because it can be seen that our data points are merged.
## we will remove Id and Species from  features(x) 
```

```python
x=data.drop(["Species","Id"],axis=1)
y=data["Species"]
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 500, n_init = 10, random_state = 0,algorithm="full")
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    

plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 200, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
print(set(y_kmeans))
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.scatter(x[y_kmeans==0]["SepalLengthCm"],x[y_kmeans==0]["SepalWidthCm"] , c='blue', label='Iris-setosa',alpha=0.3, edgecolors='none')
ax1.scatter(x[y_kmeans==1]["SepalLengthCm"],x[y_kmeans==1]["SepalWidthCm"] , c='orange', label='Iris-versicolor',alpha=0.3, edgecolors='none')
ax1.scatter(x[y_kmeans==2]["SepalLengthCm"],x[y_kmeans==2]["SepalWidthCm"] , c='green', label='Iris-virginica',alpha=0.3, edgecolors='none')
ax1.set_xlabel("SepalLengthCm")
ax1.set_ylabel("SepalWidthCm")
ax1.legend()
ax1.set_title("KMeans")
ax1.grid(True)

ax2.scatter(data[data["Species"]==0]["SepalLengthCm"],data[data["Species"]==0]["SepalWidthCm"] , c='blue', label='Iris-setosa',alpha=0.3, edgecolors='none')
ax2.scatter(data[data["Species"]==1]["SepalLengthCm"],data[data["Species"]==1]["SepalWidthCm"] , c='orange', label='Iris-versicolor',alpha=0.3, edgecolors='none')
ax2.scatter(data[data["Species"]==2]["SepalLengthCm"],data[data["Species"]==2]["SepalWidthCm"] , c='green', label='Iris-virginica',alpha=0.3, edgecolors='none')
ax2.set_xlabel("SepalLengthCm")
ax2.set_ylabel("SepalWidthCm")
ax2.legend()
ax2.grid(True)



plt.show()
```
